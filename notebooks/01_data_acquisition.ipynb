```python
# -------------------------------
# Import des librairies
# -------------------------------
import pandas as pd
import requests
from bs4 import BeautifulSoup
from sqlalchemy import create_engine

# -------------------------------
# 1. Lecture des fichiers CSV, JSON et Excel
# -------------------------------

# CSV
csv_path = "../data/student_performance.csv"
df = pd.read_csv(csv_path, sep=',')
print("Aperçu des données CSV :")
display(df.head())

print("\nInformations sur le dataset CSV :")
df.info()

# JSON
json_path = "../data/movies.json"
movies = pd.read_json(json_path)
print("\nAperçu des données JSON :")
display(movies.head())

# Excel
excel_path = "../data/DataProjet.xlsx"
DataProjet = pd.read_excel(excel_path, skiprows=5, usecols="B:H")
print("\nAperçu des données Excel :")
display(DataProjet.head())

# -------------------------------
# 2. Connexion à la base de données MySQL
# -------------------------------
engine = create_engine('mysql+pymysql://root:@localhost/school_db')
query = "SELECT * FROM student_performance WHERE `COL 5` > 50"
df_top_students = pd.read_sql(query, engine)

print("\nTop students (COL 5 > 50) :")
display(df_top_students.head())

# -------------------------------
# 3. Récupération et lecture du fichier robots.txt
# -------------------------------
base_url = "https://books.toscrape.com/"
robots_url = base_url + "robots.txt"
response_robots = requests.get(robots_url)

print("\nrobots.txt :")
if response_robots.status_code == 200:
    print(response_robots.text)
else:
    print("robots.txt : not found")

# -------------------------------
# 4. Web scraping des titres de livres
# -------------------------------
page = requests.get(base_url)
soup = BeautifulSoup(page.content, 'html.parser')
titles = soup.find_all('h3')

print("\n5 premiers titres de livres :")
for title in titles[:5]:
    book_name = title.select_one('a').get('title')
    print(f"- {book_name}")

# -------------------------------
# 5. Consommation d'une API et conversion en DataFrame
# -------------------------------
api_url = "https://jsonplaceholder.typicode.com/users"
response_api = requests.get(api_url)
data_json = response_api.json()

df_users = pd.DataFrame(data_json)
print("\nAperçu des utilisateurs de l'API :")
display(df_users[['id', 'name', 'username', 'email', 'phone']].head())
```
