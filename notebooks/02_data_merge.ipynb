# Organisation du code Python

Ce notebook organise le code Python fourni en sections logiques pour une meilleure lisibilité et gestion.

## 1. Importations nécessaires

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

## 2. Lecture des données CSV

```python
df_csv = pd.read_csv("../data/student-mat.csv")
print("Aperçu des données CSV :")
print(df_csv.head())
```

## 3. Consommation d'une API

```python
api_response = requests.get("https://jsonplaceholder.typicode.com/posts")
df_api = pd.DataFrame(api_response.json())
print("Aperçu des données de l'API :")
print(df_api.head())
```

## 4. Web Scraping

```python
url = "https://www.bbc.com/news"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
titles = [h2.text for h2 in soup.find_all("h2")]
df_web = pd.DataFrame(titles, columns=["titre"])
print("Aperçu des titres scrapés :")
print(df_web.head())
```

## 5. Fusion des données (concaténation)

```python
df_final = pd.concat([df_csv, df_api.head(len(df_csv)), df_web.head(len(df_csv))], axis=1)
print("Aperçu du DataFrame final après concaténation :")
print(df_final.head())
```
